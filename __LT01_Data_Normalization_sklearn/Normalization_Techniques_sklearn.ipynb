# **MinMaxScaler**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = [[3, 2], [15, 6], [0, 10], [1, 18]]
```

```python
scaler = MinMaxScaler()
print(scaler.fit(data))
```

```python
print(scaler.data_max_)
```

```python
print(scaler.data_min_)
```

```python
print(scaler.transform(data))
```

# **StandardScaler**

```python
from sklearn.preprocessing import StandardScaler
```

```python
data = [[3, 2], [15, 6], [0, 10], [1, 18]]
```

```python
scaler = StandardScaler()
print(scaler.fit(data))
```

```python
print(scaler.mean_)
```

```python
print(scaler.transform(data))
```

# **Log Scaler**

```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer
```

```python
data = np.array([[0,1], [2,3]])
```

```python
transformer = FunctionTransformer(np.log1p)
```

```python
print(transformer.transform(data))
```

# **MaxAbsScaler**

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
data = [[ 1., -1.,  2.],
        [ 2.,  0.,  0.],
        [ 0.,  1., -1.]]
```

```python
scaler = MaxAbsScaler()
print(scaler.fit(data))
```

```python
print(scaler.max_abs_)
```

```python
print(scaler.transform(data))
```

# **RobustScaler**

```python
from sklearn.preprocessing import RobustScaler
```

```python
data = [[ 1., -2.,  2.],
        [ -2.,  1.,  3.],
        [ 4.,  1., -2.]]
```

```python
scaler = RobustScaler()
print(scaler.fit(data))
```

```python
print(scaler.center_)
```

```python
print(scaler.transform(data))
```

# **Normalizer**

```python
from sklearn.preprocessing import Normalizer
```

```python
data = [[4, 1, 2, 2],
        [1, 3, 9, 3],
        [5, 7, 5, 1]]
```

```python
trans_l1 = Normalizer(norm='l1')
trans_l2 = Normalizer()
trans_max = Normalizer(norm='max')
print(trans_l1.fit(data))
print(trans_l2.fit(data))
print(trans_max.fit(data))
```

```python
print(trans_l1.transform(data))
```

```python
print(trans_l2.transform(data))
```

```python
print(trans_max.transform(data))
```
